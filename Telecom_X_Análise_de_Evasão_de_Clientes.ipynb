# Padroniza√ß√£o e Transforma√ß√£o de Dados

A padroniza√ß√£o e transforma√ß√£o de dados √© uma etapa opcional, mas altamente recomendada, pois visa tornar as informa√ß√µes mais consistentes, compreens√≠veis e adequadas para an√°lise. Durante essa fase, voc√™ pode, por exemplo, converter valores textuais como "Sim" e "N√£o" em valores bin√°rios (1 e 0), facilitando o processamento matem√°tico e a aplica√ß√£o de modelos anal√≠ticos.

Al√©m disso, traduzir ou renomear colunas e dados torna os dados mais acess√≠veis e compreens√≠veis, especialmente quando se lida com fontes externas ou termos t√©cnicos. Embora n√£o seja obrigat√≥ria, essa etapa pode melhorar significativamente a clareza e a comunica√ß√£o dos resultados, facilitando a interpreta√ß√£o e evitando confus√µes, especialmente ao compartilhar informa√ß√µes com stakeholders n√£o t√©cnicos.

```python
mapa_colunas = {                                                                # Dicion√°rio com os novos nomes das colunas.
    'customerID': 'id_cliente',
    'Churn': 'cancelamento',
    'customer.gender': 'genero',
    'customer.SeniorCitizen': 'idoso',
    'customer.Partner': 'possui_parceiro',
    'customer.Dependents': 'possui_dependentes',
    'customer.tenure': 'tempo_contrato_meses',
    'phone.PhoneService': 'servico_telefone',
    'phone.MultipleLines': 'multiplas_linhas',
    'internet.InternetService': 'tipo_internet',
    'internet.OnlineSecurity': 'seguranca_online',
    'internet.OnlineBackup': 'backup_online',
    'internet.DeviceProtection': 'protecao_dispositivo',
    'internet.TechSupport': 'suporte_tecnico',
    'internet.StreamingTV': 'streaming_tv',
    'internet.StreamingMovies': 'streaming_filmes',
    'account.Contract': 'tipo_contrato',
    'account.PaperlessBilling': 'fatura_digital',
    'account.PaymentMethod': 'metodo_pagamento',
    'account.Charges.Monthly': 'valor_mensal',
    'account.Charges.Total': 'total_cobrado'
}
df = df.rename(columns=mapa_colunas)
```

```python
df.head()
```

```python
colunas_binarias = [                                                            # Lista de colunas bin√°rias para convers√£o (Sim/N√£o ou Yes/No ‚Üí 1/0).
    'cancelamento',
    'possui_parceiro',
    'possui_dependentes',
    'servico_telefone',
    'multiplas_linhas',
    'protecao_dispositivo',
    'suporte_tecnico',
    'streaming_tv',
    'streaming_filmes',
    'fatura_digital'
]
for col in colunas_binarias:                                                    # Percorre cada coluna da lista para realizar a convers√£o.
    df[col] = (                                                                 # Aplica a convers√£o.
        df[col]                                                                 # Acessa a coluna.
        .str.strip()                                                            # Remove espa√ßos em branco.
        .str.lower()                                                            # Converte o texto para letras min√∫sculas.
        .map({'yes': 1, 'no': 0, 'sim': 1, 'n√£o': 0})                           # Mapeia os valores categ√≥ricos para num√©ricos.
    )
```
